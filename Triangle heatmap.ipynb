```python
# trangle heatmap with

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

corr = np.corrcoef(np.random.randn(10, 200)) # random data for practise
#corr = tempdata.corr()
sns.set_theme(style="ticks", font_scale=1.4, font = "Arial")

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True


with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(10, 8))
    
    ax = sns.heatmap(
        corr,
        mask=mask,
        annot=True,
        vmin=-1, vmax=1, center=0,
        cmap="RdBu_r",
        square=True,
        linewidths=0.5,

    )
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
