```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Data = pd.read_csv('resource/PK_AS_Concentration.csv')
Data.head()
```

```python
sns.set_theme(style="white")

sns.relplot(x="Salt Concentration_M", y="Protein Concentration_mg_mL", hue="Crystal", size="Crystal Size",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=Data)
plt.show()
```
