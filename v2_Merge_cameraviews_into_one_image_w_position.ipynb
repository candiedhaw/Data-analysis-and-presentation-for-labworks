```python

# create matrix to store all flattened images
immatrix = array([array(Image.open("cropped_cameraview/0800 011422_Meas "+str(im)+".JPG")).flatten()
                  for im in imlist],'f')
```

```python
# create empty matrix
plt.style.use('dark_background')
fig, axs = plt.subplots(ncols=16, nrows=6)
fig.set_figwidth(35)
fig.set_figheight(15)
gray()

#remove the ticks
for ax in axs.flat:
    ax.set(xticks=[],yticks=[])

#remove white space between subplots
plt.subplots_adjust(wspace=0,hspace=0)
```

```python
#subplot the images to correlated position
def subplot_img(idx):
    r= (idx-1)%col #col idx(starts from 0)
    q=(idx-1)//col #row id(starts from 0)
    axs[q,r].imshow(immatrix[idx-1].reshape(m,n))
```

```python
for i in imlist:
    subplot_img(i)
fig
```
