```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('resource/ProteinaseK_CRT_6features 091421.csv')
df.head()
```

```python
# only keep the #crystals related df 
df = df[['Experiment Index','Ammonium Sulfate Concentration_M','Proteinase K Concentration_mg_mL',
         'CRT Crystal Production',
         'MG8 Crystal Production',
         'MG6 Crystal Production',
         'MCC6 Crystal Production',
         'MCS6 Crystal Production',
         'MCT3 Crystal Production', 
         'MAST Crystal Production'
        ]]


# Filter: AS @1.8M, Proteinase K @3mg/mL
df = df[df['Ammonium Sulfate Concentration_M']== 1.8] 
df = df[df['Proteinase K Concentration_mg_mL'].astype(float) ==3]

df.head()
```

```python
df = df.set_index('Experiment Index').drop(['Ammonium Sulfate Concentration_M',
                                            'Proteinase K Concentration_mg_mL'],
                                           axis=1)

df = df.rename({'CRT Crystal Production': 'CTRL Crystal Production'}, axis=1)
df.head()
```

```python
# one column is string, convert to float
a = df['MCS6 Crystal Production'].astype(float)
df['MCS6 Crystal Production'] = a
```

```python
# split columns name
coumns_striped = [s.split()[0] for s in list(df)]
df.columns = coumns_striped
df.head()
```

```python
df = df[df['CTRL'].notna()]
```

```python
#gradient background only apply to non-special index, remove experiment index
df = df.reset_index(drop=True)
```

```python
#gradient background
cm = sns.light_palette("green", as_cmap=True)

df_gradient = df.style.background_gradient(cmap=cm, axis=None)
df_gradient
```
