```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('resource/PK_Crystal_size - PK24 cons.csv').dropna(axis=1)
df.head()
```

```python

sns.color_palette("Set2")
sns.set(rc={'figure.figsize':(12,7)})


ax = sns.violinplot(x='Protein_conc_mg/mL', y="Crystal_D_um", data=df,
                    palette = "pastel",
                    width = 1,
                    scale = 'area')

ax.set(xlabel="Protein Concentration (mg/mL)", ylabel = "Crystal Dimention(um)")
```
