```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('resource/ProteinaseK_CRT_6features 091421.csv')
df.head()
```

```python
df = df[['Experiment Index','Ammonium Sulfate Concentration_M','Proteinase K Concentration_mg_mL',
         'CRT Onset Time_hrs','MG8 Onset Time_hrs','MG6 Onset Time_hrs','MCC6 Onset Time_hrs',
         'MCS6 Onset Time_hrs','MCT3 Onset Time_hrs', 'MAST Onset Time_hrs']]
df = df[df['Ammonium Sulfate Concentration_M']== 1.8] 
df = df[df['Proteinase K Concentration_mg_mL'].astype(float) ==3]
#df = df[df['Proteinase K Concentration_mg_mL'].astype(float) >=2]
df.head()
```

```python
ctrl_onset = df['CRT Onset Time_hrs'].dropna().astype(float)
mg8_onset = df['MG8 Onset Time_hrs'].dropna().astype(float)
mcc6_onset = df['MCC6 Onset Time_hrs'].dropna().astype(float)
mg6_onset = df['MG6 Onset Time_hrs'].dropna().astype(float)
mcs6_onset = df['MCS6 Onset Time_hrs'].dropna().astype(float)
mast_onset = df['MAST Onset Time_hrs'].dropna().astype(float)
mct3_onset = df['MCT3 Onset Time_hrs'].dropna().astype(float)
```

```python
#onset time(hrs)
df_onset=pd.DataFrame({'CTRL': ctrl_onset,
                       'MG8': mg8_onset,
                       'MG6': mg6_onset,
                       'MCC6': mcc6_onset,
                       'MCS6': mcs6_onset,
                       'MCT3': mct3_onset,
                       'MAST': mast_onset
                       
                       })
df_onset.index = np.arange(1, len(df_onset)+1)
df_onset = df_onset.drop(8)
df_onset.head()
```

```python
cm = sns.light_palette("green", as_cmap=True)

df_onset = df_onset.style.background_gradient(cmap=cm, axis=None)
df_onset
```

```python
df_onset.to_excel('onset background gradient.xlsx')
```

## Method1: average out-->compare

```python
df = pd.DataFrame(df_onset.mean(axis=0).round(1),columns=['mean'])
df = df.swapaxes("index", "columns")
```

```python
# time improvement %
df_timprove1=pd.DataFrame({
                       'MG8': (-(df['MG8']-df['CTRL'])/df['CTRL']*100).round(1),
                       'MCC6': (-(df['MCC6']-df['CTRL'])/df['CTRL']*100).round(1),
                       'MG6': (-(df['MG6']-df['CTRL'])/df['CTRL']*100).round(1),
                       'MCS6': (-(df['MCS6']-df['CTRL'])/df['CTRL']*100).round(1),
                       'MAST': (-(df['MAST']-df['CTRL'])/df['CTRL']*100).round(1),
                       'MCT3': (-(df['MCT3']-df['CTRL'])/df['CTRL']*100).round(1)
                       })
df_timprove1
```

```python
df_onset_mean = pd.DataFrame((df_onset.mean(axis=0).round(1).astype(str)+' ('+df_onset.std(axis=0).round(1).astype(str)+')'),columns=['mean'])
df_onset_mean.swapaxes("index", "columns")
```

## Method2: compare--> average out

```python

df_timprove=pd.DataFrame({'MG8': (-(mg8_onset-ctrl_onset)/ctrl_onset*100),
                          'MG6': (-(mg6_onset-ctrl_onset)/ctrl_onset*100),
                          'MCC6': (-(mcc6_onset-ctrl_onset)/ctrl_onset*100),
                          'MCS6': (-(mcs6_onset-ctrl_onset)/ctrl_onset*100),
                          'MCT3': (-(mct3_onset-ctrl_onset)/ctrl_onset*100),
                          'MAST': (-(mast_onset-ctrl_onset)/ctrl_onset*100)
                         })
df_timprove = df_timprove.drop(63)
df_timprove.head()
```

```python
df_timprove_mean = pd.DataFrame((df_timprove.mean(axis=0).round(1).astype(str)+' ('+df_timprove.std(axis=0).round(1).astype(str)+')'),columns=['mean'])
df_timprove_mean.swapaxes("index", "columns")
```
