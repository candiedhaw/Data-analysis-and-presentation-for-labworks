```python
import pandas as pd
import numpy as np

#f_df: feature position df (384)
#v_df: volumn position df (depend on experimental setting)
#df: DLS result df
```

```python
f_df = pd.read_csv('resource/384_plate_features.csv', sep=',', header = None)
f_df.head()
```

```python
#convert index 0-->A
idx= f_df.index.to_series().apply(lambda x: chr(ord('a') + x)).str.upper()
f_df.index = idx
f_df.head()
```

```python
#column name = column name +1
col = f_df.columns+1
f_df.columns = col
f_df.head()
```

```python
# melt matrix to series
f_df = f_df.reset_index().melt(id_vars='index').sort_values(['index','variable'])
f_df.index = f_df['index'] +f_df['variable'].astype(str)
f_df = f_df.drop(['index','variable'], axis=1)
f_df = f_df.rename({'value': 'Feature'}, axis=1)
f_df.head()
```

```python
v_df = pd.read_csv('resource/12212021_volume.csv', sep=',', header = None)
v_df.head()
```

```python
v_df.index= v_df.index.to_series().apply(lambda x: chr(ord('a') + x)).str.upper()
v_df.columns = v_df.columns+1
v_df.head()
```

```python
(row, col) = v_df.shape
col# column, the number used for df reshape later
```

```python
v_df = v_df.reset_index().melt(id_vars='index').sort_values(['index','variable'])
v_df.index = v_df['index'] +v_df['variable'].astype(str)
v_df.head()

v_df = v_df.drop(['index','variable'], axis=1).rename({'value': 'Volumn (uL)'}, axis=1)
v_df.head()
```

```python
df = pd.read_csv('resource/DLS_20211221_result.csv', sep=',', header = None)
df.head()
```

```python
# first raw-->column name
header = df.iloc[0]
df = df[1:]
df.columns = header
df.head()
```

```python
# to get the plate reading sequence: left to right, top to bottom
idx= df.index.astype(int).to_series()
idx.head()
```

```python
#Add the plate position information to the df
#5%2 #remainder
#5//2 #quotient
r = (idx-1)%col+1
q = (idx-1)//col

print(r)
print(q)

#0-->A...
q = q.apply(lambda x: chr(ord('a') + x)).str.upper() 
df['row']=q
df['column']=r.astype(str)
df.head()
```

```python
df.index = df['row']+df['column']
df = df.drop(['row','column'], axis=1)
df.head()
```

```python
#merge volumn, feature, DLS result to df
merge_df = pd.merge(pd.merge(v_df, f_df, how='inner', left_index=True, right_index=True),df,how='inner', left_index=True, right_index=True)
merge_df.head()
```

```python
# Drop useless columns, sort, save as
df = merge_df[['Volumn (uL)','Feature','Radius (nm)']]
df = df.sort_values(['Volumn (uL)','Feature']) 
df.to_excel('DLS_20211221_sorted_rawdata.xlsx')
df.head(10)
```
