```python
df = pd.read_csv('resource/plasma cleanning 092221_3.csv', sep=' ', header = None)
df.head()
```

```python
para = df[4].str.replace('\(','', regex = True).str.split('_') #slip the volume, test,and feature

df['Volume'] = para.apply(lambda x: x[0][0])
df['Test'] = para.apply(lambda x: x[1])
df['Feature'] = para.apply(lambda x: x[2])
df['Slide'] = para.apply(lambda x: x[3][0])
df.head()
```

```python
df['Contact Angle (°)'] = df[6].astype(float)
df['Volume'] = df['Volume'].astype(int)
df.head()
```

```python
df = df.drop([0,1,2,3,4,5,6],axis = 1)
df.fillna(0)
df.head()
```

```python
#Volume = 1 or 2
#df = df[df['Volume']==1]
df = df[df['Volume']==2]
df.head()
```

```python
df = df.sort_values(['Volume','Test','Feature','Slide']) 
df.head(10)
```

```python
df_d1 = df[df['Test'] == 'DAY1']# time-dependant study only
df = df[df['Test'] != 'DAY1']# filter out time-dependant result
df_d1
```

```python
# time-dependant study only
df_ctrl = df[df['Feature']=='CTRL']
df_d0 = df_ctrl[(df_ctrl['Test']=='C6')|(df_ctrl['Test']=='E4')]
df_og = df_ctrl[(df_ctrl['Test']=='A1')|(df_ctrl['Test']=='D2')]
df_d0
```

```python
pivot = df_og.groupby(['Volume','Test','Feature','Slide']).agg({'Contact Angle (°)':(np.nanmean, np.nanstd)}).round(1)
pivot = pivot.fillna(0)
pivot
```

   
   
reshape pivot table to regular table

```python
pivot.columns = pivot.columns.droplevel(0)
pivot = pivot.reset_index().rename_axis(None, axis=1)
pivot
```

Remove control result for tablewise

```python
pivot['Contact Angle (°)'] = pivot['nanmean'].astype(str)+' ('+pivot['nanstd'].astype(str)+')'
pivot.head()
```

```python
pivot = pivot.pivot(index=['Test','Slide'],
                    columns='Feature',
                    values='Contact Angle (°)')
pivot
```
